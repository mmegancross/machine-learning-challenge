```python
import pandas as pd
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
for col in df:
    print(col)
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values. 
selected_features = df[['koi_fpflag_nt','koi_fpflag_ss','koi_fpflag_co','koi_fpflag_ec',
                        'koi_period','koi_period_err1','koi_period_err2',
                        'koi_time0bk','koi_time0bk_err1','koi_time0bk_err2',
                        'koi_impact','koi_impact_err1','koi_impact_err2',
                        'koi_duration','koi_duration_err1','koi_duration_err2',
                        'koi_depth','koi_depth_err1','koi_depth_err2',
                        'koi_prad','koi_prad_err1','koi_prad_err2',
                        'koi_teq','koi_insol','koi_insol_err1','koi_insol_err2',
                        'koi_model_snr','koi_steff','koi_steff_err1','koi_steff_err2',
                        'koi_slogg','koi_slogg_err1','koi_slogg_err2',
                        'koi_srad','koi_srad_err1','koi_srad_err2',
                        'ra','dec','koi_kepmag']]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df["koi_disposition"]
X = selected_features
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from tensorflow.keras.utils import to_categorical

#label encoding 
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
y_train_encoded = label_encoder.transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data

X_scale = MinMaxScaler().fit(X_train)
y_scale = MinMaxScaler().fit(X_train)
```

```python
X_train_scaled = X_scale.transform(X_train)
X_test_scaled = X_scale.transform(X_test)
```

```python
X_test_scaled
```

# Train the Model



```python
# Create the Logistic Regression Model
from sklearn.linear_model import LogisticRegression

model2 = LogisticRegression()
model2.fit(X_train_scaled, y_train_encoded)
```

```python
print(f"Training Data Score: {model2.score(X_train_scaled, y_train_encoded)}")
print(f"Testing Data Score: {model2.score(X_test_scaled, y_test_encoded)}")
```

```python
#Classification report

from sklearn.metrics import classification_report
predictions = model2.predict(X_test_scaled)
print(classification_report(y_test_encoded, predictions))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {
    "C": [1, 5, 10, 50],
    "penalty": ['l1', 'l2']
}

grid = GridSearchCV(model2, param_grid, verbose=3)
```

```python
#get params 
grid.get_params().keys()

# Train the model with GridSearch
grid.fit(X_train_scaled, y_train_encoded)
```

```python
#List best parameters and best score
print(grid.best_params_)
print(grid.best_score_)
```

```python
#Make predictions with hypertuned model
predictions = grid.predict(X_test_scaled)
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test_encoded, predictions))
```
